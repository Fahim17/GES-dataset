# rename and stitch images

```python
# Rename
import os

# Path to your image folder
folder_path = 'C:\\Users\\ChinoMo\\Downloads\\CVUSA_drone'

# Get all image files, sorted by name
image_files = sorted([f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])

# Rename them to 1.ext, 2.ext, ..., 5.ext
for i, filename in enumerate(image_files, start=1):
    ext = os.path.splitext(filename)[1]
    new_name = f"{i}{ext}"
    old_path = os.path.join(folder_path, filename)
    new_path = os.path.join(folder_path, new_name)
    os.rename(old_path, new_path)
    print(f"Renamed: {filename} -> {new_name}")
```

```python
# Stitch
import os
from PIL import Image

# Path to the main folder containing subfolders
main_folder = 'C:\\Users\\ChinoMo\\Downloads\\CVUSA_drone\\vigor'
output_folder = 'stitched_output'
os.makedirs(output_folder, exist_ok=True)

# Get all subfolders
subfolders = sorted([
    os.path.join(main_folder, d) for d in os.listdir(main_folder)
    if os.path.isdir(os.path.join(main_folder, d))
])

# Stitch images for each location (1 to 5)
for loc in range(1, 6):
    images = []
    for folder in subfolders:
        img_path = os.path.join(folder, f'{loc}.jpeg')
        if os.path.exists(img_path):
            img = Image.open(img_path)
            images.append(img)
        else:
            print(f"Warning: Missing {img_path}")

    if images:
        # Resize all images to same height
        min_height = min(img.height for img in images)
        resized = [img.resize((int(img.width * min_height / img.height), min_height)) for img in images]

        # Calculate total width
        total_width = sum(img.width for img in resized)

        # Create blank canvas
        stitched = Image.new('RGB', (total_width, min_height))

        # Paste images side by side
        x_offset = 0
        for img in resized:
            stitched.paste(img, (x_offset, 0))
            x_offset += img.width

        # Save
        stitched.save(os.path.join(output_folder, f'location_{loc}.jpg'))
        print(f"Saved: location_{loc}.jpg")
```

# Elevation


```python
#Open Elevation

import pandas as pd
import requests
import time
from tqdm import tqdm

# Load your CSV
df = pd.read_csv('vigor_temp_coordinates.csv')  # Must have 'latitude' and 'longitude' columns

def get_elevation(lat, lon):
    url = 'https://api.open-elevation.com/api/v1/lookup'
    params = {'locations': f'{lat},{lon}'}
    try:
        response = requests.get(url, params=params)
        if response.status_code == 200:
            return response.json()['results'][0]['elevation']
        else:
            print(f"Error {response.status_code} for {lat},{lon}")
            return None
    except Exception as e:
        print(f"Request failed: {e}")
        return None

# Add elevation column
elevations = []
for index, row in tqdm(df.iterrows()):
    lat = row['latitude']
    lon = row['longitude']
    elevation = get_elevation(lat, lon)
    elevations.append(elevation)
    time.sleep(0.5)  # Respect rate limit


df['elevation'] = elevations
df.to_csv('vigor_temp_with_elevation.csv', index=False)

```
